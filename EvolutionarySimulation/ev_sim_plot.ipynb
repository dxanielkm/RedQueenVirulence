```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV
data = pd.read_csv('virulence_distribution.csv')

# Option 1: pivot to a 2D matrix: rows=Step, cols=Alpha
pivoted = data.pivot(index='Step', columns='Alpha', values='Population')

# Plot a heatmap
plt.figure(figsize=(8,6))
# step is the index (y-axis), alpha is the columns (x-axis)
# square rooted to make it more visible 
plt.imshow(np.power(pivoted.T, 1/3), aspect='auto', origin='lower',
           extent=[pivoted.index.min(), pivoted.index.max(),
                   pivoted.columns.min(), pivoted.columns.max()],
           cmap='gray_r')  # or any other colormap

plt.colorbar(label='Population Density')
plt.xlabel('Evolutionary Step')
plt.ylabel('Virulence')
plt.title('Population Distribution of Virulence Over Evolutionary Time')
plt.show()
```
